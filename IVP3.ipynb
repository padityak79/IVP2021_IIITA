```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

drive.mount('/content/drive')


img_unp = cv2.imread('/content/input.png', 0)
m, n = img_unp.shape
cv2_imshow(img_unp)

img = np.zeros([m+2, n+2])

for i in range(1, m+2):
    for j in range(1, n+2):
      if(i == 0 or j == 0 or i == 1 or j == 1 or i == m-1 or j == n-1 or i == m or j == n):
        img[i,j] = 0
      else:
        img[i,j] = img_unp[i-2,j-2]
mask = np.ones([5, 5], dtype = int)
mask = mask / 25

img_new = np.zeros([m, n])

for i in range(1, m-2):
    for j in range(1, n-2):
        temp = img[i-2, j-2]*mask[0, 0]+img[i-2, j-1]*mask[0, 1]+img[i-2, j]*mask[0, 2]+img[i-2, j+1]*mask[0, 3]+ img[i-2, j+2]*mask[0, 4]+img[i-1, j - 2]*mask[1, 0]+img[i - 1, j-1]*mask[1, 1]+img[i - 1, j]*mask[1, 2]+img[i - 1, j + 1]*mask[1, 3]
        + img[i - 1, j+2]*mask[1, 4] + img[i, j-2]*mask[2, 0] + img[i, j-1]*mask[2, 1] + img[i, j]*mask[2, 2] + img[i, j+1]*mask[2, 3] + img[i, j+2]*mask[2, 4] + img[i+1, j-2]*mask[3, 0]
        + img[i+1, j-1]*mask[3, 1] + img[i+1, j]*mask[3, 2] + img[i+1, j+1]*mask[3, 3] + img[i+1, j+2]*mask[3, 4] + img[i+2, j-2]*mask[4, 0] + img[i+2, j-1]*mask[4, 1] + img[i+2, j]*mask[4, 2]
        + img[i+2, j+1]*mask[4, 3] + img[i+2, j+2]*mask[4, 4]
         
        img_new[i, j]= temp
img_new = img_new.astype(np.uint8)
cv2_imshow(img_new)
```
